```python
from tkinter import *
import tkinter.messagebox
window = Tk()
#title
window.title("To_Do_List :")

#to hold the listbox and scrollbar
frame_task = Frame(window)
frame_task.pack()
#to hold items in listbox
list_box = Listbox(frame_task,bg="black",fg="white",height=15,width=50,font="bold")
list_box.pack(side=tkinter.LEFT)
#scroll bar to get exceed 
scroll_bar = Scrollbar(frame_task)
scroll_bar.pack(side=RIGHT,fill=tkinter.Y)
list_box.config(yscrollcommand=scroll_bar.set)
scroll_bar.config(command=list_box.yview)

# entry buttons
entry_button = Button(window,text="Add Task",width=50,command="entertask")
entry_button.pack(pady=3)

# delete button
delete_button = Button(window,text="Delete Task",width=50,command="deletetask")
delete_button.pack(pady=3)

# mark button
mark_button = Button(window,text="Mark as completed",width=50,command="markcompleted")
mark_button.pack(pady=3)

window.mainloop()

def entertask():
    input_text =""
    def add():
        input_text=entry_task.get(1.0, "end-1c")
        if input_text =="":
            tkinter.messagebox.showwarning(title="warning!",message="please enter text")
        else:
            list_box.insert(END, input_text)
            root1.destroy()
        root1=Tk()
        root1.title("Add task")
        entry_task=Text(root1,width=40,height=4)
        entry_task.pack()
        button_temp=Button(root1,text="Add task",command=add)
        button_temp.pack()
        root1.mainloop()
def deletetask():
    #selects the selected item and then deletes it 
    selected=list_box.curselection()
    list_box.delete(selected[0])
def markcompleted():
    marked=list_box.curselection()
    temp=marked[0]
    #store the text of selected item in a string
    temp_marked=list_box.get(marked)
    #update it 
    temp_marked=temp_marked+" ✔"
    #delete it then insert it 
    list_box.delete(temp)
    list_box.insert(temp,temp_marked)
        
```
